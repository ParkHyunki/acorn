## KNN 알고리즘

    1) 특징
        + 이해하기 쉽고 직관적인 모델
        + 더 복잡한 알고리즘을 적용하기 전에 시도해볼 수 있는 알고리즘
        + 훈련셋이 너무 크면 예측이 느려진다(데이터가 너무 많으면 예측이 느려진다)
        + 주도적 알고리즘으로 사용x
        + 새로 입력된 데이터들을 어떻게 분류할 것인가 : 거리 측정 통해 가장 가까운 이웃을 찾아 분류
        
    2) Parameter : 정확도와 연관
        + 이웃의 개수(k)
        + 데이터 포인트 사이의 거리 재는 방법 : L2 norm
            - L1 norm : 전체 거리에 대한 비용의 절대값의 합계 (장애물 피해 갈 수 있는 거리의 총합)
            - L2 norm : 전체 거리에 대한 비용의 합계 (가장 빠르게 갈 수 있는 거리 합)

```python
### 알고리즘 확인

mglearn.plots.plot_knn_classification(n_neighbors=3)  #> 알고리즘 설명 그래프(n_neighbors=이웃 지정할 때는 홀수로 지정)
```

```python
### 머신러닝 구현 : 훈련용 데이터와 테스트용 데이터를 따로 구분하여 학습시킨다

from sklearn.model_selection import train_test_split  #> 훈련용과 테스트용 구분

X, y = mglearn.datasets.make_forge()
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)  #> 무작위로 섞어서 데이터 추출(난수값 고정)
```

```python
X_test
```

```python
### 학습(훈련) 시키기

from sklearn.neighbors import KNeighborsClassifier

clf = KNeighborsClassifier(n_neighbors=3)  #> n_neighbors=3 : hyper parameter(사용자가 파라미터를 직접 입력)

clf.fit(X_train, y_train)  #> 기존 데이터로 훈련
```

```python
### 학습(훈련) 결과 테스트

print("테스트셋으로 예측 : {}".format(clf.predict(X_test)))
```

```python
### 테스트 정확도 측정

print("테스트 결과 정확도 : {:.2f}".format(clf.score(X_test, y_test)))  
    #> {:.2f}소수점 둘째자리까지 표시. score(예측값, 결과값) 몇 개 맞았고 틀렸는지 평균 내어 확인
```

```python
### 분석

fig, axes = plt.subplots(1, 3, figsize=(10, 3))

for n_neighbors, ax in zip([1, 3, 9], axes):
    clf = KNeighborsClassifier(n_neighbors=n_neighbors).fit(X, y)
    mglearn.plots.plot_2d_separator(clf, X, ax=ax, fill=True, alpha=.4, eps=0.5)
    mglearn.discrete_scatter(X[:, 0], X[:, 1], y, ax=ax)
    ax.set_title("{} 이웃".format(n_neighbors))
    ax.set_xlabel("특성 0")
    ax.set_ylabel("특성 1")
    
axes[0].legend(loc=3)

    #> 이웃이 적을수록 복잡도는 높아진다(현재 주어진 데이터를 정확히 구분)
    #>        -> 과대 적합: 훈련에는 적합하나 새로운 데이터 통한 예측x
    #> 이웃이 많아질수록 구분선 단순(오류 발생) -> 새로운 데이터에 대한 예측은 더 유리할 수 있다(일반화+단순 모델)
    #>        -> 과소 적합: 너무 단순하여 분류 의미x
```

```python
### 적정 파라미터 설정 위해 그래프로 교차점 확인

from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state=0, stratify=cancer.target)  
    #> random에 따라 교차점 달라진다. stratify= 비율에 맡게 훈련용 데이터와 테스트용 데이터 분류
    #> 공통 : 이웃 수 많을수록 훈련 정확도는 낮아지고 테스트 정확도는 높아진다
    
training_accuracy = []
test_accuracy = []  #> 훈련용과 테스트용 비교하여 정확도 측정 위함

neighbors_settings = range(1, 11)  #> 최적의 파라미터 확인 위해 이웃 수 변경(1~10)

for n_neighbors in neighbors_settings:
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(X_train, y_train)
    
    training_accuracy.append(clf.score(X_train, y_train))
    test_accuracy.append(clf.score(X_test, y_test))
    
plt.plot(neighbors_settings, training_accuracy, label="훈련 정확도")
plt.plot(neighbors_settings, test_accuracy, label="테스트 정확도")
plt.ylabel("정확도")
plt.xlabel("n_neighbors")
plt.legend()
```

```python
from sklearn.neighbors import KNeighborsRegressor

X, y = mglearn.datasets.make_wave(n_samples=40)

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

reg = KNeighborsRegressor(n_neighbors=3)
reg.fit(X_train, y_train)
```

```python
print("테스트셋으로 예측 : {}".format(reg.predict(X_test)))
```

```python
print("테스트 결과 정확도 : {:.2f}".format(reg.score(X_test, y_test)))  
```

```python
### 분석

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

line = np.linspace(-3, 3, 1000).reshape(-1, 1)  #> 눈금 범위 설정
for n_neighbors, ax in zip([1, 3, 9], axes):
    reg = KNeighborsRegressor(n_neighbors=n_neighbors).fit(X_train, y_train)
    
    ax.plot(line, reg.predict(line))
    ax.plot(X_train, y_train, "^", c=mglearn.cm2(0), markersize=8)
    ax.plot(X_test, y_test, "v", c=mglearn.cm2(1), markersize=8)
    
    train = reg.score(X_train, y_train)
    test = reg.score(X_test, y_test)
    ax.set_title("{}이웃의 훈련 점수:{:.2f}  테스트 점수 : {:.2f}".format(n_neighbors, train, test))
    ax.set_xlabel("특성")
    ax.set_ylabel("타겟")
```

```python
### 영화 분류 프로그램
import operator

# 데이터셋 생성
def createDataSet():
    group = np.array([[1.0, 2.0], [1.0, 4.0], [4.0, 1.0], [4.0, 2.0]])
    label = ["Action", "Action", "Romance", "Romance"]
    return group, label

# 거리 계산하는 함수 생성(L2 norm 공식 그대로 따라간다)
def calcDistance(new_x, data):
    dataSize = data.shape[0]  #> 데이터 개수 추출(4)
    diffMat = np.tile(new_x, (dataSize, 1)) - data  #> 새로운 값의 거리 계산(행렬 맞춰서 계산 편리하게 하기 위함)
    sqDiffMat = diffMat**2
    sqDistance = sqDiffMat.sum(axis=1)
    
    distance = sqDistance**0.5
    sortedDistance = distance.argsort()  #> 순서에 해당하는 위치 출력
    return sortedDistance

def findMajorityClass(new_x, target, k, sortedDistances):
    classCount = {}
    
    for i in range(k):
        voteLabel = target[sortedDistances[i]]
        classCount[voteLabel] = classCount.get(voteLabel, 0) + 1
        
    return sorted(classCount.items(), reverse=True, key=operator.itemgetter(1))
        
    

def classify(new_x, data, target, k):  #> KNN 분류
    # 거리 계산
    sortedDistances = calcDistance(new_x, data)
    sortedDistanceCount = findMajorityClass(new_x, target, k, sortedDistances)
    
    return sortedDistanceCount[0][0]
```

```python
X, y = createDataSet()
result = classify([2.0, 3.0], X, y, 3)
print("결과는 {} 영화이다".format(result))
```
