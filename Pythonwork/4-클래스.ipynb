# 특징

    1) 비슷한 것끼리의 묶음(block)
        - 변수(상태, state): class 내 소속되어 있는 변수는 명칭이 다름
                            (멤버 변수, 필드, 속성, 인스턴스 변수)
        - 함수(행동, behavior: 모듈화의 최소단위): class 내 소속되어 있는 함수는 명칭이 다름
                              (멤버 함수, 메서드)
        
    2) 규모가 큰 프로그램의 유지/보수를 쉽게 하기 위해 사용
    
    3) 여러 개의 함수를 묶어서 좀더 큰 단위의 모듈로 관리
    
    4) 설계도(추상화) 역할: 메모리나 프로그램에 영향을 주지 않음(계획)
        - 객체: 클래스의 실체화
        
    5) 사용자 정의 자료형(내장 자료형으로 묶을 수 없는 자료형들)
    
    6) 객체(Object, 분리/독립적으로 기능 가능)의 조건(객체 지향 언어) 
        - 상속성(Ingeritance): 상위 class의 특성이 하위 class에도 적용됨(재사용성)
        - 다형성(Polymoriphism): 객체가 다양한 모양을 갖는다(편의)
        - 캡슐화(Encapsulation): 모듈화. 추상화. 블랙박스. class로 묶인 것 자체가 하나의 캡슐(편리, 보안성)
        
    7) 객체 설계 5원칙(SOLID)
        - 단일 책임 원칙(SRP): 클래스는 하나의 기능
        - 개방-폐쇄 원칙(OCP): 요구사항 수정에도 기존 구성요소는 변경x / 기존 구성요소 확장하여 재사용 가능
        - 리스코프치환 원칙(LSP): 서브 타입은 언제나 기반 타입으로 교체 가능
        - 의존 역전 원칙(DIP): 하위 모듈의 변경이 상위 모듈의 변경 요구x
        - 인터페이스 분리 원칙(ISP): 클래스가 다른 클래스에 종속될 때는 최소한의 인터페이스만 사용

# 작성 문법

    1) self
        - 현재 사용 중인 객체의 주소를 저장하고 있는 참조 변수
        - read only
        
    2) 생성자(__init__)와 소멸자(__del__)
        - 메서드
        - callback: 함수를 직접 호출하는 것이 아닌, 시스템이 필요할 때 호출해서 사용
        - 생성자: 객체 생성 용도
            >> 객체를 처음 만들 때 한 번만 호출(초기화 용도로 사용)
        - 소멸자: 객체가 사라질 때 호출됨
        
    3) static
        - 동일 공간을 모든 객체들이 공동으로 사용
        - 변수 차원에서는 class 이름으로 접근하여 사용(but, 객체에서 먼저 사용중일 경우 static 적용x)
        - 메서드 차원에서는 @staticmethod 라는 decorator 사용
        
    4) _ or __
        > _: 기본 인터프리터에 사용
            >> 가장 최근에 사용된 값 저장
            >> 값 무시
        > _, __: 변수명과 같이 사용(접근 제한)  ex> _변수, _메서드, __변수, __메서드
            >> _변수명: 가급적 변경 금지 표시(강제성x)
            >> __변수명: 변경 금지 표시 강조(숨김. 강제성)
            
    5) 상속(부모class -> 자식class)
        - 표준화 작업 가능
        - 설계시 유의점
            >> 부모class는 최소한의 코드로 작성(자식class들이 필요로 하는 교집합 코드만 마련)
            >> 중복 코드x
            >> 상속받은 자식class는 반드시 기능을 확장시켜야 한다
        - 상속관계 모델링 언어: UML
        - Overriding(재정의): 부모class로부터 상속받은 기능을 수정하여 사용  -> 다형성 특징(구현)
        - Overloading(중복정의): 같은 이름으로 여러 함수를 만들 수 있다  -> 다형성 특징. but, python에서 지원x(*default인수로 대체)
        - super(): 부모의 주소 알려줌
        - overriding 이용하여 표준화 구현
        - 다중 상속: 부모class 기능의 중복 문제. 부모class 기능 추적 불가능  but, 기능적으로는 더 유리
        - 추상(abstract) class
            >> 기본적으로 추상 메서드를 한 개 이상 가지고 있어야 한다
            >> 추상 메서드: 내용이 없는 메서드
                            반복적으로 자식class가 메서드 수정하여 사용해야 하는 경우 사용
                            강제적 overriding 필요(실수 사전 방지)
            >> 객체를 만들지 않는다(only 자식class 상속)
        - object: 모든 class 상속
        
        
# 클래스의 관계

    1) is a 관계(A is a B): 상속 관계(B로부터 A가 상속을 받았나?)
        - 특정 클래스 객체가 다른 클래스 객체에게 자신의 멤버를 물려주는 상속 관계
        - 강한 결합 관계: 서로 의존 관계
        - UML 삼각형
    
    2) has a 관계(A has a B): 포함 관계(A는 B를 포함?)
        - 특정 클래스 내에서 해당 클래스 멤버처럼 다른 클래스 객체를 포함하고 있는 경우
            >> Aggregation(약한 포함 관계)
            >> Composition(강한 포함 관계)
        - UML 마름모

```python
class Person:
    nation = "한국"
    name = str()
    age = int()
    height = float()
    
    def info(self):
        print("국적: ", self.nation)
        print("이름: ", self.name)
        print("나이: ", self.age)
        print("신장: ", self.height)
        print("-" * 20)
    
    @staticmethod     # 객체별로 따로 가져갈 필요가 없는 메서드 static
    def tel(arg):     # 주소 구분할 필요가 없기 때문에 self 사용x
        print("대표 전화:{}({})".format(arg, Person.nation))     # self 대신 class 이름 사용
        
#####

obj3 = Person()
obj3.name = "장돌식"
obj3.age = 22
obj3.height = 179
obj3.info()

obj3.nation = "고구려"    
Person.nation = "대한민국"     # 변수 차원 static

obj3.info()

Person.tel("02-111-1111")     # 메서드 차원 static 사용
```

```python
class Employee:
    cnt = 0
    
    def __init__(self, num, name, Salary, Hobby):
        self.num = num
        self.name = name
        self.Salary = Salary
        self.Hobby = Hobby
        Employee.cnt += 1
            
    def display(self):
        print("사번: ", self.num)
        print("이름: ", self.name)
        print("급여: ", self.Salary)
        print("취미: ", end = " ")
        for i in self.Hobby:
            print(i, end = " ")
        print("\n", "*" * 20)
        
    def setSalary(self, arg):
        self.Salary = arg
        
    def appendHobby(self, arg):
        for i in arg:
            self.Hobby.append(i)
            
    @staticmethod
    def getCount():
        return Employee.cnt
                
#####
    
e1 = Employee(1111, "홍길동", 1000000, ["여행", "낚시", "운동"])
e2 = Employee(1112, "임꺽정", 1200000, ["자전거", "볼링"])
e3 = Employee(1113, "장돌석", 1500000, ["독서"])

e1.display()
e2.display()

e3.setSalary(2500000)
e3.appendHobby(["피아노", "드론"])
e3.display()

total = Employee.getCount()
print("총 인원수: ", total)
```

```python
# 사칙연산 계산기 프로그램 class에 다른 class 추가

class Calculator:
    def __init__(self, n1=None, n2=None):     # 입력값 초기화 위함
        self.first = n1
        self.second = n2
        
    def setData(self, n1=None, n2=None):     # 입력값 수정 위함
        self.first = n1
        self.second = n2
        
    def sum(self):
        result = self.first + self.second
        return result
    
    def sub(self):
        result = self.first - self.second
        return result
    
    def mul(self):
        result = self.first * self.second
        return result
    
    def div(self):
        result = self.first / self.second
        return result
    
#####

class Calculator_v1(Calculator):     # 부모class Calculator 상속
    def pow(self):
        result = self.first ** self.second
        return result

#####

calc1 = Calculator()
calc1.setData(10, 20)
print(calc1.sum())

calc2 = Calculator(50, 20)
print(calc2.mul())

calc3 = Calculator(50, 10)
print(calc3.div())

calc4 = Calculator_v1(5, 20)
print(calc4.div())
print(calc4.pow())
```

```python
# overriding(재정의): div 함수 수정

class Calculator:
    def __init__(self, n1=None, n2=None):     # 입력값 초기화 위함
        self.first = n1
        self.second = n2
        
    def setData(self, n1=None, n2=None):     # 입력값 수정 위함
        self.first = n1
        self.second = n2
        
    def sum(self):
        result = self.first + self.second
        return result
    
    def sub(self):
        result = self.first - self.second
        return result
    
    def mul(self):
        result = self.first * self.second
        return result
    
    def div(self):
        result = self.first / self.second
        return result
    
#####

class Calculator_v1(Calculator):     # 부모class Calculator 상속
    def pow(self):
        result = self.first ** self.second
        return result
    
    def div(self):     # div 함수 overriding: 모양 똑같이 명시 후 수정 내용 
        if self.second == 0:
            return "0 으로 나눌 수 없습니다"
            
        else:
            return self.first / self.second
    
#####

calc = Calculator_v1(10, 0)
print(calc.div())
```

```python
# super()

class Person:
    say = "난 사람이야"
    age = "30"
    
    def __init__(self, nai):
        print("Person 생성자")
        self.age = nai
        
    def info(self):
        print("나이: {}, 하고 싶은 말: {}".format(self.age, self.say))
        
#####

class Employee(Person):
    say = "일하는 동물"
    subject = "근로자"     # 부모class와 자식class에 같은 명의 변수 존재(overriding)
    
    def __init__(self):
        print("Employee 생성자")
        
    def einfo(self):
        print(self.age)
        print(super().age)     # 부모class의 age값 접근 가능
        
#####

e1 = Employee()
e1.info()    # 부모class의 age + 자식class say
#e2.Employee(20)     # Error: 생성자(__init__)는 상속되지 않는다
e1.age = 20     # 자식class에서 부모class로부터 상속받은 변수 수정 가능
e1.info()

e1.einfo()
```

```python
# static: @staticmethod, @classmethod .. 단독 class에서 사용 방법의 차이: 형식만 차이가 있고 기능 동일

class Test:
    num = 10
    
    @staticmethod     # 공통으로 사용할 수 있는 유틸리티
    def calc(x):
        return x + 10 + Test.num     # self 쓸 수 있으므로 반드시 class 명으로 접근
    
#####

class Test1:
    num = 10
    
    @classmethod     
    def calc(cls, x):     # class 이름 전달받아야 하는 자리에 cls
        return x + 10 + cls.num     # class 명 맞출 필요 없이 cls로 가능(staticmethod보다 편리)

#####

print(Test.calc(50))     # 객체 사용하지 않고 바로 출력 가능
print(Test1.calc(30))
```

```python
# overriding을 이용한 표준화(인터페이스)

class Volume:
    vol = int()
    def volumeUp(self, vol):
        self.vol += vol
        print("볼륨의 크기는 %d만큼 증가하여 현재 볼륨은 %d입니다" %(vol, self.vol))
    def volumeDown(self, vol):
        self.vol -= vol
        print("볼륨의 크기는 %d만큼 감소하여 현재 볼륨은 %d입니다" %(vol, self.vol))
        
#####

class TV(Volume):
    def volumeUp(self, vol):
        self.vol += vol
        print("TV 볼륨의 크기는 %d만큼 증가하여 현재 볼륨은 %d입니다" %(vol, self.vol))
    def volumeDown(self, vol):
        self.vol -= vol
        print("TV 볼륨의 크기는 %d만큼 감소하여 현재 볼륨은 %d입니다" %(vol, self.vol))
        
#####

class Radio(Volume):
    def volumeUp(self, vol):
        self.vol += vol
        print("라디오 볼륨의 크기는 %d만큼 증가하여 현재 볼륨은 %d입니다" %(vol, self.vol))
    def volumeDown(self, vol):
        self.vol -= vol
        print("라디오 볼륨의 크기는 %d만큼 감소하여 현재 볼륨은 %d입니다" %(vol, self.vol))

#####

class Speaker(Volume):
    def volumeUp(self, vol):
        self.vol += vol
        print("스피커 볼륨의 크기는 %d만큼 증가하여 현재 볼륨은 %d입니다" %(vol, self.vol))
    def volumeDown(self, vol):
        self.vol -= vol
        print("스피커 볼륨의 크기는 %d만큼 감소하여 현재 볼륨은 %d입니다" %(vol, self.vol))

#####

tv = TV()
radio = Radio()
speaker = Speaker()

tv.volumeUp(5)
tv.volumeDown(2)

radio.volumeUp(1)
radio.volumeDown(6)

speaker.volumeUp(10)
speaker.volumeDown(3)
```

```python
# 강한 포함 관계
# p145 예제

class Handle:
    quantity = 0
    
    def leftTurn(self, q):
        self.quantity = q
        return "좌회전"
    
    def rightTurn(self, q):
        self.quantity = q
        return "우회전"
    
    def straight(self, q):
        self.quantity = q
        return "직진"

#####

class Car:
    turnShow = "정지"
    
    def __init__(self, ownerName):
        self.ownerName = ownerName
        self.handle = Handle()     # Handle() 클래스의 객체를 통해 좌회전/우회전 판단 -> 강한 포함 관계
        
    def turnHandle(self, q):
        if q > 0:
            self.turnShow = self.handle.rightTurn(q)
        elif q < 0:
            self.turnShow = self.handle.leftTurn(q)
        elif q == 0:
            self.turnShow = self.handle.straight(q)
            
#####

tom = Car("tom")
tom.turnHandle(10)
print(tom.ownerName + "의 회전량은",  tom.turnShow, tom.handle.quantity)
```

```python
# 약한 포함 관계: 클래스 관계는 약할수록 좋다
# p147 예제

class Fridge:
    isOpened = False
    foods = []
    
    def open(self):
        self.isOpened = True
        print("냉장고 문 열기")
        
    def close(self):
        self.isOpened = False
        print("냉장고 문 닫기")
        
    def put(self, arg):
        if self.isOpened == True:
            self.foods.append(arg)
            print("냉장고 속에 음식이 저장되었습니다")
            self.list()
        else:
            print("냉장고 문이 닫혀있어 음식을 넣을 수 없습니다")
            
    def list(self):
        for f in self.foods:
            print("-", f.name, f.ex_data)
        print()

#####

class FoodData:
    def __init__(self, name, ex_data):
        self.name = name
        self.ex_data = ex_data
        
#####

f = Fridge()
apple = FoodData("사과", "2019.8.20.")
banana = FoodData("바나나", "2019.8.6.")

f.open()
f.put(apple)
f.put(banana)
```
